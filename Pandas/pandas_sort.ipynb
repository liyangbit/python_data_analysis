```python
import pandas as pd
import numpy as np
print(f'pandas version: {pd.__version__}')
```

```python
data = {
    'brand':['Python数据之道','价值前瞻','菜鸟数据之道','Python','Java'],
    'B':[4,6,8,12,10],
    'A':[10,2,5,20,16],
    'D':[6,18,14,6,12],
    'years':[4,1,1,30,30],
    'C':[8,12,18,8,2],
}

index = [9,3,4,5,2]

df = pd.DataFrame(data=data,index=index)
df
```

## sort_index

### 按索引名称排序

```python
# sort_index

df.sort_index()
```

```python
# sort_index

df.sort_index(ascending=False)
```

### 按列的名称排序

通过设置参数 `axis=1` 可实现按列的名称排序

```python
# 按列的名称排序

df.sort_index(axis=1)
```

```python
# 按列的名称排序
df.sort_index(axis=1,ascending=False)
```

## sort_values

### 按单个列的值排序

```python
# 按单个列的值排序，默认是升序
df.sort_values('B')
```

```python
# 按单个列的值排序
df.sort_values('B',ascending=False)
```

### 按多个列的值排序

```python
# 按多个列的值排序，默认是升序
df.sort_values(['years','B'])
```

```python
# 按多个列的值排序
# 'years' 升序， 'B' 降序
df.sort_values(['years','B'],ascending=[True,False])
```

### 选择排序算法

选择排序算法，参数 kind 默认是 'quicksort'，其他算法有 mergesort, heapsort, stable。

该参数只针对单个列时才有效。

```python
# 只能针对单个列
df.sort_values('brand',kind='quicksort')
```

```python
# 选择排序算法，默认是 'quicksort'，
#其他算法有 mergesort, heapsort, stable
# 只能针对单个列
df.sort_values('brand',kind='mergesort')
```

```python
# 选择排序算法，默认是 'quicksort'，其他算法有 mergesort, heapsort, stable
# 只能针对单个列
df.sort_values('brand',kind='heapsort')
```

### 忽略索引

```python
# 忽略索引
df.sort_values('B',ignore_index=True)
```

### inplace

`inplace = True`：不创建新的对象，直接对原始对象进行修改； 默认是 `False`，即创建新的对象进行修改，原对象不变，和深复制和浅复制有些类似。

```python
# inplace

df.sort_values('B',inplace=True)
df
```

### 缺失值

```python
data = {
    'brand':['Python数据之道','价值前瞻','菜鸟数据之道','Python','Java'],
    'B':[4,6,8,np.nan,12],
    'A':['Lemon','emma','ZW','app','John'],
    'D':[6,18,14,6,12],
    'years':[4,1,1,30,30],
    'C':[8,12,18,8,2],
}

index = [9,3,4,5,2]

df1 = pd.DataFrame(data=data,index=index)
df1
```

```python
# 包含缺失值，NA值的排序

# NA值排在最前面
df1.sort_values('B',na_position='first')
```

```python
# 包含缺失值，NA值的排序

# NA值排在最后面
df1.sort_values('B',na_position='last')
```

### key 参数

通过设置 key 参数，可以将列按照特定条件进行排序

```python
df1.sort_values('A')
```

```python
df1.sort_values('A',key=lambda col:col.str.lower())
```
